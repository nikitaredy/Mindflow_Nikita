```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
data=pd.read_csv('C:\\Users\\niico\\Downloads\\householdtask3.csv')
```

```python
display(data.head(10))
```

```python
#scatter plot with year against own
plt.scatter(data['year'],data['own'])

#title
plt.title("Scatter plot")

#setting the x and y labels
plt.xlabel('year')
plt.ylabel('own')

plt.show()
```

```python
#Line chart with year against own
plt.plot(data['year'],data['own'])
plt.title("Line chart")
plt.xlabel('year')
plt.ylabel('own')

plt.show()
```

```python
#barchart
plt.bar(data['year'],data['own'])
plt.title("BAR CHART")
plt.xlabel('year')
plt.ylabel('own')

plt.show()
```

```python
#histogram
plt.hist(data['income'])
plt.title("Histogram")
plt.xlabel('year')
plt.ylabel('own')

plt.show()
```
